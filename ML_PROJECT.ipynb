```python
import pandas as pd

# Define a class for natural disasters
class NaturalDisaster:
    def __init__(self, entity, year, economic_damage_usd):
        self.entity = entity
        self.year = year
        self.economic_damage_usd = economic_damage_usd

    def __str__(self):
        return f"Entity: {self.entity}, Year: {self.year}, Economic Damage (US$): {self.economic_damage_usd}"

# Load the dataset into a Pandas DataFrame
data = pd.read_csv('/content/economic-damage-from-natural-disasters.csv')  # Replace 'natural_disasters.csv' with your dataset file

# Create a list of NaturalDisaster objects from the dataset
disasters_list = []
for index, row in data.iterrows():
    disaster = NaturalDisaster(row['Entity'], row['Year'], row['Total economic damage from natural disasters (US$)'])
    disasters_list.append(disaster)

# Example: Display the information for each disaster
for disaster in disasters_list:
    print(disaster)
```
